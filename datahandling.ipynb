# Data Cleaning
Dataset may contain null/nan values and duplicates which can cost more time or can cause errors due to lack of information. So cleaning of data is an important task in data science.

```python
dataset = Pd.read_csv("/content/sample_data/movies.csv")
```

```python
#a = dataset.groupby(by="Station.State")
dataset
```

### Removing Duplicates

```python
ds_red = dataset.groupby(by="MOVIES").agg("count")

dfsx = dataset.groupby(by="MOVIES")['MOVIES'].count()
print(dfsx)

print()

print("No of Distinct Rows = ", len(ds_red))
print("Difference = ",abs(len(ds_red) - len(dataset)) )
```

The fact that grouping them based on the **name of movies** varies in length from the original dataset proves that there are around 3182 movie reviews which are **duplicates**

```python
reduced_ds = dataset.drop_duplicates("MOVIES",keep="first")
```

Notice that the duplicate reviews of movie "The Imperfects" is removed. similarly other duplicates will also be removed from dataset based on the given column "Movies".

You can see the great reduction in the count of total dataset, compared to the amount of data which are distinct.

```python
print(len(dataset))
print(len(reduced_ds))
```

### Handling missing data
We have two columns with major missing data : "Gross" and "Rating"

```python
#Shallow Copy
msh_ds = Pd.DataFrame(reduced_ds)
msh_ds = msh_ds.reset_index()

# Notice that Gross is not a necessary attribute, so we can drop it
msh_ds['Gross']
```

```python
msh_ds = msh_ds.drop('Gross',axis=1)
msh_ds
```

### Handling the Missing values in necessary rows
Time to replace the NaN's with default values

```python
msh_ds.describe()
```

```python
# Since the avg is 6.63
msh_ds['RATING'] = msh_ds['RATING'].fillna(value=6.0)
msh_ds
```
