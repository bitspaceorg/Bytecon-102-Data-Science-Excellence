# **BAR GRAPH**

```python
import matplotlib.pyplot as plt
import numpy as np
# importing required modules
```

```python
x = np.array(["C","CPP","PY","LUA","JAVA","JS"])

y = np.array([60,90,98,35,75,85])

plt.bar(x,y)

plt.show()
```

```python
plt.bar(x,y)

plt.title("Most Used Languages",fontsize="24")

plt.xlabel("Languages",fontsize="12")

plt.ylabel("No.of.people",fontsize="16")

plt.show()
```

```python
plt.bar(x,y,color="hotpink",width=0.5)

plt.title("Most Used Languages")

plt.xlabel("Languages")

plt.ylabel("No.of.people")

plt.show()
```

```python
plt.bar(x,y,color=["blue","hotpink","yellow","blue","orange","yellow"],width=0.5)

plt.title("Most Used Languages")

plt.xlabel("Languages")

plt.ylabel("No.of.people")

plt.show()
```

```python
plt.barh(x,y,color=["blue","hotpink","yellow","blue","orange","yellow"],height=0.5)

plt.title("Most Used Languages")

plt.ylabel("Languages")

plt.xlabel("No.of.people")

plt.show()
```

# **Pie Chart**

```python
import matplotlib.pyplot as plt
import numpy as np
# importing required modules
```

```python

x = np.array([60,90,98,35,75,85])

plt.pie(x)

plt.show()
```

```python
label = np.array(["C","CPP","PY","LUA","JAVA","JS"])

x = np.array([60,90,98,35,75,85])

plt.pie(x,labels=label)

plt.show()
```

```python
label = np.array(["C","CPP","PY","LUA","JAVA","JS"])

x = np.array([60,90,98,35,75,85])

plt.pie(x,startangle=90,labels=label)

plt.show()
```

```python
label = np.array(["C","CPP","PY","LUA","JAVA","JS"])

x = np.array([60,90,98,35,75,85])

plt.pie(x,startangle=90,labels=label,explode=[0,0,0,0,0,0.2])

plt.show()
```

```python
label = np.array(["C","CPP","PY","LUA","JAVA","JS"])

x = np.array([60,90,98,35,75,85])

plt.pie(x,labels=label,shadow=True,explode=[0,0,0,0,0,0.2])

plt.show()
```

```python
label = np.array(["C","CPP","PY","LUA","JAVA","JS"])

x = np.array([60,90,98,35,75,85])

plt.pie(x,labels=label,shadow=True,explode=[0,0,0,0,0,0.2])

plt.legend(title="langs")

plt.show()
```

# **Sub Plots**

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
y = np.array([3, 9, 7, 10])

plt.subplot(1, 2, 1)
          # r  c  p
plt.plot(y)

#plot 2:
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(y)

plt.show()
```

```python
y = np.array([3, 9, 7, 10])

plt.subplot(1, 2, 1)
          # r  c  p
plt.plot(y)

plt.title("hello")

#plot 2:
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(y)
plt.title("world")

plt.suptitle("super title")

plt.show()
```

# **Scatter**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y,color="red")

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y,color="black")

plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([6,7,2,4,2])
c = ["red","black","green","blue","pink"]

plt.scatter(x,y,color=c)

plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([6,7,2,4,2])
c = [10,20,30,40,50]

plt.scatter(x,y,c=c,cmap='Accent') # Blues , Dark2

plt.colorbar()

plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([6,7,2,4,2])
s = [1000,200,13,2600,20]

plt.scatter(x,y,s=s)

plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
```

```python
x = np.random.randint(50, size=(50))
y = np.random.randint(50, size=(50))
colors = np.random.randint(50, size=(50))
sizes = 10 * np.random.randint(50, size=(50))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='Dark2')

plt.colorbar()

plt.show()
```
